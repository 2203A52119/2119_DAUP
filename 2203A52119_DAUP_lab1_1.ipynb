```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Step 1: Creating vectors, factors, lists, matrices, and data frames
# Vectors
data = pd.read_csv('/content/insurance.csv')
ages = data['age'].values
smokers = data['smoker'].values

# Factors (using pandas categories)
data['smoker'] = pd.Categorical(data['smoker'])
smoker_factor = data['smoker']

# Lists
list_data = [ages[:5], smokers[:5], data['expenses'][:5].values]

# Matrix (reshaping 'age' and 'bmi' columns)
matrix = data[['age', 'bmi']].head(5).to_numpy()

# Data frame (pandas DataFrame is already in use)
df_sample = data.head(5)

# Display created structures
structures = {
    "Vector (Ages)": ages[:5],
    "Factor (Smoker)": smoker_factor[:5].tolist(),
    "List (Sample Data)": list_data,
    "Matrix (Age and BMI)": matrix,
    "Data Frame (Sample)": df_sample,
}
structures
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['smoker'].map({'yes': 1, 'no': 0}), c='blue', alpha=0.6)
plt.title('Scatter Plot: Age vs. Smoking', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Smoking Status (0=Non-Smoker, 1=Smoker)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 2. Scatter Plot: Ages 20-40 vs. Smoking
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['age'], filtered_data['smoker'].map({'yes': 1, 'no': 0}), c='green', alpha=0.6)
plt.title('Scatter Plot: Ages 20-40 vs. Smoking', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Smoking Status (0=Non-Smoker, 1=Smoker)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
smoker_expenses = data.groupby('smoker')['expenses'].mean()
plt.figure(figsize=(8, 6))
smoker_expenses.plot(kind='bar', color=['red', 'green'])
plt.title('Average Expenses of Smokers vs. Non-Smokers', fontsize=14)
plt.xlabel('Smoking Status', fontsize=12)
plt.ylabel('Average Expenses', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Non-Smoker', 'Smoker'])
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
gender_smoker_count = data[data['smoker'] == 'yes'].groupby('sex').size()
plt.figure(figsize=(8, 6))
gender_smoker_count.plot(kind='bar', color=['blue', 'orange'])
plt.title('Number of Smokers by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count of Smokers', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['expenses'], c='purple', alpha=0.6)
plt.title('Scatter Plot: Age vs. Expenses', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Expenses', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
