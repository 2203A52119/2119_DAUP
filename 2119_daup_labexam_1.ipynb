```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/insurance.csv")
numerical_cols = df.select_dtypes(include=np.number).columns

stats_df = pd.DataFrame(index=numerical_cols)
stats_df['Mean'] = df[numerical_cols].mean()
stats_df['Median'] = df[numerical_cols].median()
stats_df['Variance'] = df[numerical_cols].var()
stats_df['Standard Deviation'] = df[numerical_cols].std()
stats_df['Skewness'] = df[numerical_cols].skew()
stats_df['Kurtosis'] = df[numerical_cols].kurt()

print(stats_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='children', y='expenses', data=df)
plt.title('Boxplot of Expenses vs. Children')
plt.xlabel('Children')
plt.ylabel('Expenses')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='children', y='age', data=df)
plt.title('Boxplot of Age vs. Children')
plt.xlabel('Children')
plt.ylabel('Age')
plt.show()
```
